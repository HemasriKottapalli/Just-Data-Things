```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('V Kohli vs RG Sharma Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'],color='red')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'blue')

plt.title('V Kohli vs RG Sharma Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'],color='red', linestyle='dashed')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'blue', linestyle='dotted')

plt.title('V Kohli vs RG Sharma Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'],color='red', linestyle='solid',linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'blue', linestyle='dotted')

plt.title('V Kohli vs RG Sharma Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
#showing plot points
plt.plot(batsman['index'], batsman['V Kohli'],color='red', linestyle='solid', marker='D', markersize=10)
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'blue', linestyle='dotted')

plt.title('V Kohli vs RG Sharma Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
#giving name to each line
plt.plot(batsman['index'], batsman['V Kohli'],color='red', linestyle='dashed', label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'blue', linestyle='dotted', label='Rohit')

plt.title('V Kohli vs RG Sharma Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs')
#the labels won't show on screen unless we call the legend function
plt.legend()
# plt.legend(loc='upper right')
```

```python
df = pd.read_csv('batsman_season_record.csv')
df.head()
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
#Handling multiple bar graphs beautifully
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2, color='red')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='pink')
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
plt.hist(arr)
plt.show()
```

In the above graph, there is data from 10 to 15 but it's not showing because other bins have very high data so that 10-15 bin is not even showing.
For this ww can use logarthemic scale instead of linear scale

```python
plt.hist(arr,bins=[0,10,20,30,40,50,60,70],log=True)
plt.show()
```

Pie Chart<br>
Univarient and Bivarient<br>
Numerical and Categorical <br>
Use Case : To find contribution on a standard scale

```python
data = [90,90,92,45,23]
subjects = ['maths', 'science', 'telugu','Social', 'Physics']
plt.pie(data, labels=subjects)
plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
df.head()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct ='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct ='%0.1f%%', colors=['red','pink','green','orange','black','yellow'])
plt.show()
```

```python
#exlpode shadow
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct ='%0.1f%%', explode=[0,0.1,0,1,0,0], shadow=True)
plt.show()
```
