```python
lst = []
for i in range(1,100):
    url = "https://api.themoviedb.org/3/tv/top_rated?language=en-US&page={}".format(i)
    response = requests.get(url.format(i), headers=headers)
    temp_df = pd.DataFrame(response.json()["results"])[['id', 'name','original_name', 'overview', 'first_air_date', 'popularity', 'vote_average', 'vote_count','adult','genre_ids','original_language']]
    lst.append(temp_df)
df_new = pd.concat(lst, ignore_index=True)
out = pd.concat([df, df_new])
```
