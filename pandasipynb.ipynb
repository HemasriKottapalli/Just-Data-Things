```python
def get_city_match_count(city):
  mask = data['city'] == city
  return data[mask].shape[0]
```

```python
# sorting based on city
data.sort_values('city',ascending=False)
```

```python
#Usually when the output gets printed without explicitly printing it like the above one, the changes are temporary
#To make permenant changes we have to pass inplace = true. which is pretty dangerous
data.sort_values('city',ascending=False, inplace=True) # doesn't print anything
```
